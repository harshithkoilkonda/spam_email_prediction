```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('spam.csv', sep=',', encoding='latin-1')
```

```python
data.head()
```

```python
data['v1'].value_counts()
```

```python
data.isnull().sum()
```

```python
data.shape
```

```python
ham = data[data['v1'] == 'ham']
ham.shape
```

```python
spam = data[data['v1'] == 'spam']
spam.shape
```

```python
ham = ham.sample(spam.shape[0])
ham.shape
```

```python
data = pd.concat([ham, spam], axis = 0, ignore_index = True)
data.shape
```

```python
data.sample(10)
```

DATA PREPARATION FOR TRAINING

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import classification_report
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
X_train, X_test, y_train, y_test = train_test_split(data['v2'], data['v1'],
                                                    test_size = 0.2, random_state = 0,
                                                    shuffle = True, stratify = data['v1'])
```

```python
clf = Pipeline([('tfidf', TfidfVectorizer()),
               ('rfc', RandomForestClassifier(n_estimators = 100, n_jobs = -1))])

clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
clf.predict(['you have won lottery ticket worth $2000, please click here to claim',
            'hi, how are you doing today?'])
```
